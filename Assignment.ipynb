```python
for multiple in range(1,50):
    if multiple % 3 ==0:
        print("fizz")
    elif multiple % 5 ==0:
        print("buzz")
    elif multiple % 3 ==0 and multiple % 5 ==0:
        print("fizzbuzz")
    else:
        print(multiple)
```

```python
for multiple in range(1,50):
    if multiple % 3 ==0 and multiple % 5 ==0:
        print("fizzbuzz")
    elif multiple % 5 ==0:
        print("buzz")
    elif multiple % 3 ==0:
        print("fizz")
    else:
        print(multiple)
```
